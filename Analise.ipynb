DICIONARIO:
Arquivo understat.com.csv (dados agregados por time e temporada):

league: A liga de futebol (ex: La Liga, Premier League, etc.).

year: O ano da temporada.

position: A posição final do time na tabela.

team: O nome do time.

matches: O número de partidas jogadas.

wins: Vitórias.

draws: Empates.

loses: Derrotas.

scored: Gols marcados.

missed: Gols sofridos.

pts: Pontos totais.

xG (Expected Goals): Gols esperados. É uma medida da qualidade das chances de gol criadas por um time. Um chute de dentro da pequena área tem um xG alto, enquanto um chute de muito longe tem um xG baixo.

xGA (Expected Goals Against): Gols esperados contra. Mede a qualidade das chances de gol que um time cede ao adversário.

npxG (Non-penalty Expected Goals): Gols esperados sem contar os pênaltis.

npxGA (Non-penalty Expected Goals Against): Gols esperados contra sem contar os pênaltis.

deep: Número de passes que o time completou que terminaram a 20 jardas ou menos do gol adversário.

ppda_coef: Passes por Ação Defensiva. Mede a intensidade da pressão de um time. Um número baixo significa que o time pressiona mais o adversário.

xpts (Expected Points): Pontos esperados, calculados com base nas métricas de xG e xGA.

Arquivo understat_per_game.csv (dados por jogo):

Este arquivo tem muitas das mesmas colunas que o anterior, mas os valores são por partida. As colunas mais importantes que se somam aqui são:

h_a: Indica se o time jogou em casa ('h') ou fora ('a').

result: O resultado da partida para o time em questão ('w' para vitória, 'd' para empate, 'l' para derrota).

date: A data da partida.

```python
import pandas as pd
import plotly.express as px

# Carrega o arquivo de dados
df = pd.read_csv('understat.com.csv')

# Calcula a métrica de saldo de finalização para todos os times
df['saldo_finalizacao'] = df['scored'] - df['xG']

# --- Parâmetro de Análise ---
ano_selecionado = 2018
# ---------------------------

# Pega a lista de todas as ligas únicas do arquivo
lista_de_ligas = df['league'].unique()
print(f"Gerando gráficos de Eficiência Ofensiva para: {lista_de_ligas}")

# Inicia o loop que passará por cada liga
for liga in lista_de_ligas:
    
    # Filtra os dados para a liga e ano atuais do loop
    df_liga = df[(df['league'] == liga) & (df['year'] == ano_selecionado)].copy()

    # Se não houver dados para essa combinação, pula para a próxima liga
    if df_liga.empty:
        continue
        
    # Arredonda os valores para 2 casas decimais ANTES de plotar
    df_liga['xG'] = df_liga['xG'].round(2)
    df_liga['saldo_finalizacao'] = df_liga['saldo_finalizacao'].round(2)

    # Cria o gráfico de dispersão com os dados já formatados
    fig = px.scatter(
        df_liga,
        x='xG',
        y='scored',
        color='saldo_finalizacao',
        color_continuous_scale=px.colors.diverging.RdBu,
        hover_data=['team', 'position', 'saldo_finalizacao'],
        title=f'Eficiência Ofensiva - {liga} {ano_selecionado}',
        labels={'scored': 'Gols Marcados', 'xG': 'Gols Esperados (xG)'}
    )

    # Adiciona a linha de referência y=x para dar contexto
    max_val = max(df_liga['xG'].max(), df_liga['scored'].max()) + 5
    fig.add_shape(type='line', x0=0, y0=0, x1=max_val, y1=max_val,
                  line=dict(color='Gray', width=2, dash='dash'))

    fig.update_layout(height=600)
    
    # Exibe o gráfico para a liga atual
    print(f"\n--- {liga} ---")
    fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Carrega o arquivo de dados
df = pd.read_csv('understat.com.csv')

# Calcula a métrica de saldo defensivo
df['saldo_defensivo'] = df['missed'] - df['xGA']

# --- Parâmetro de Análise ---
ano_selecionado = 2018
# ---------------------------

# Pega a lista de todas as ligas únicas do arquivo
lista_de_ligas = df['league'].unique()
print(f"Gerando gráficos de Eficiência Defensiva para: {lista_de_ligas}")

# Inicia o loop que passará por cada liga
for liga in lista_de_ligas:

    # Filtra os dados para a liga e ano atuais do loop
    df_liga = df[(df['league'] == liga) & (df['year'] == ano_selecionado)].copy()

    # Se não houver dados, pula para a próxima liga
    if df_liga.empty:
        continue
    
    # Arredonda os valores para 2 casas decimais
    df_liga['xGA'] = df_liga['xGA'].round(2)
    df_liga['saldo_defensivo'] = df_liga['saldo_defensivo'].round(2)

    # Cria o gráfico de dispersão
    fig = px.scatter(
        df_liga,
        x='xGA',
        y='missed',
        color='saldo_defensivo',
        color_continuous_scale=px.colors.diverging.RdBu_r, # Escala de cor invertida
        hover_data=['team', 'position', 'saldo_defensivo'],
        title=f'Eficiência Defensiva - {liga} {ano_selecionado}',
        labels={'missed': 'Gols Sofridos', 'xGA': 'Gols Esperados Contra (xGA)'}
    )

    # Adiciona a linha de referência y=x
    max_val = max(df_liga['xGA'].max(), df_liga['missed'].max()) + 5
    fig.add_shape(type='line', x0=0, y0=0, x1=max_val, y1=max_val,
                  line=dict(color='Gray', width=2, dash='dash'))

    fig.update_layout(height=600)

    # Exibe o gráfico para a liga atual
    print(f"\n--- {liga} ---")
    fig.show()
```

```python
# Investigue a correlação da coluna pts com as demais métricas (xG, xGA, ppda_coef, deep, etc.). Quero um ranking claro dos fatores que mais impactam a pontuação final.
import pandas as pd
import plotly.express as px

# Carrega o arquivo de dados
df = pd.read_csv('understat.com.csv')

# 1. Selecionamos as colunas numéricas relevantes
colunas_numericas = [
    'wins', 'draws', 'loses', 'scored', 'missed', 'pts', 
    'xG', 'npxG', 'xGA', 'npxGA', 'ppda_coef', 'deep', 'xpts'
]
df_numerico = df[colunas_numericas]

# 2. Calculamos a matriz de correlação
matriz_corr = df_numerico.corr()

# 3. Pegamos apenas as correlações com 'pts' e ordenamos
corr_com_pts = matriz_corr['pts'].sort_values(ascending=False)

# 4. Removemos a correlação de 'pts' com ela mesma e formatamos a tabela
corr_com_pts = corr_com_pts.drop('pts').reset_index()
corr_com_pts.columns = ['Métrica', 'Correlação']

# Correlação foi para 2 casas decimais
corr_com_pts['Correlação'] = corr_com_pts['Correlação'].round(2)
# ---------------------------

# 5. Criamos o gráfico de barras ordenado com os dados já formatados
fig = px.bar(
    corr_com_pts,
    x='Métrica',
    y='Correlação',
    color='Correlação',
    text='Correlação', # Adiciona o valor arredondado como texto em cima da barra
    color_continuous_scale=px.colors.diverging.RdBu,
    color_continuous_midpoint=0,
    title='Ranking de Fatores: O que mais se correlaciona com Pontos (pts)?',
    labels={'Correlação': 'Coeficiente de Correlação com Pontos'}
)

fig.update_traces(textposition='outside') # Posição do texto
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('understat.com.csv')
df['diferenca_xpts'] = df['pts'] - df['xpts']

ano_selecionado = 2017

# 2. Pega a lista de todas as ligas únicas do dataframe
lista_de_ligas = df['league'].unique()
print(f"Ligas encontradas no arquivo: {lista_de_ligas}")

# 3. Inicia o loop 'for' para iterar sobre cada liga
for liga in lista_de_ligas:
    print(f"\n--- Gerando gráfico para a liga: {liga} ({ano_selecionado}) ---")

    # Filtra os dados para a liga ATUAL do loop e o ano selecionado
    df_temporada = df[(df['league'] == liga) & (df['year'] == ano_selecionado)].copy()

    # Ordena os times da liga atual pela diferença de performance
    df_temporada = df_temporada.sort_values(by='diferenca_xpts', ascending=False)

    # Cria a figura do gráfico de barras
    fig = px.bar(
        df_temporada,
        x='team',
        y='diferenca_xpts',
        color='diferenca_xpts',
        color_continuous_scale=px.colors.diverging.RdBu,
        hover_data={
            'position': True,  
            'pts': True,      
            'xpts': ':.2f',    
            'diferenca_xpts': ':.2f' 
        },
        title=f'Performance Real vs. Esperada - {liga} {ano_selecionado}',
        labels={
            'team': 'Time',
            'diferenca_xpts': 'Diferença (Pontos Reais - Pontos Esperados)',
            'position': 'Posição Final',
            'pts': 'Pontos Reais',
            'xpts': 'Pontos Esperados'
        }
    )

    fig.add_hline(y=0, line_width=2, line_dash="dash", line_color="black")
    fig.show()
```
